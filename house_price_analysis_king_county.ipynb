# House Price Analysis & Prediction – King County, USA

## Project Objective
Analyze residential house sales data from King County to identify key factors affecting house prices and build regression models to predict prices accurately.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression, Ridge
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error

%matplotlib inline
```

## 1. Data Loading & Initial Inspection


```python
df = pd.read_csv("housing.csv")
df.head()
```

```python
df.dtypes
```

```python
df.info()

```

```python
df.describe()
```

### Purpose
- Understand dataset structure  
- Identify missing values  
- Check data types and value ranges  


## 2. Data Cleaning & Preprocessing


```python
df.drop(['id', 'Unnamed: 0'], axis=1, inplace=True)
```

```python
df[['bedrooms', 'bathrooms']].isnull().sum()
```

```python
df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].mean())
df['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].mean())
```

```python
df[['bedrooms', 'bathrooms']].isnull().sum()
```

```python
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

### Why
- Identifier columns add no predictive value  
- Mean imputation preserves dataset size while handling missing data  


## 3. Exploratory Data Analysis (EDA)

## 4. Feature Selection & Baseline Models


### Target Variable


```python
Y = df['price']
```

### Baseline Model 1: Longitude Only


```python
X = df[['long']]
lm = LinearRegression()
lm.fit(X, Y)
lm.score(X, Y)
```

### Baseline Model 2: Sqft Living


```python
X = df[['sqft_living']]
lm.fit(X, Y)
lm.score(X, Y)
```

### Multivariate Linear Regression


```python
X.isnull().sum().sort_values(ascending=False)
```

```python
features = [
    "floors", "waterfront", "lat", "bedrooms", "sqft_basement",
    "view", "bathrooms", "sqft_living15", "sqft_above",
    "grade", "sqft_living"
]

X = df[features]
Y = df['price']

lm = LinearRegression()
lm.fit(X, Y)
lm.score(X, Y)
```

### Observation
Multivariate models significantly outperform single-feature baselines.


## 5. Polynomial Regression Using Pipeline


```python
pipe = Pipeline([
    ('scale', StandardScaler()),
    ('poly', PolynomialFeatures(include_bias=False)),
    ('model', LinearRegression())
])

pipe.fit(X, Y)
pipe.score(X, Y)
```

### Why
- Captures non-linear relationships  
- Pipeline ensures clean and repeatable preprocessing  


## 6. Model Evaluation with Train–Test Split


```python
x_train, x_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.15, random_state=1
)
```

## 7. Ridge Regression & Regularization


### Ridge Regression (Linear)


```python
ridge = Ridge(alpha=0.1)
ridge.fit(x_train, y_train)
ridge.score(x_test, y_test)
```

### Ridge Regression with Polynomial Features


```python
pr = PolynomialFeatures(degree=2)

x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.transform(x_test)

ridge.fit(x_train_pr, y_train)
ridge.score(x_test_pr, y_test)
```

## 8. Proper Model Evaluation


```python
y_pred = ridge.predict(x_test_pr)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
mae = mean_absolute_error(y_test, y_pred)

rmse, mae
```

## 9. Model Comparison Summary


```python
model_scores = {
    "Linear (Sqft Living)": LinearRegression().fit(df[['sqft_living']], Y)
                                .score(df[['sqft_living']], Y),
    "Multivariate Linear": lm.fit(X, Y).score(X, Y),
    "Polynomial + Ridge (Test)": ridge.score(x_test_pr, y_test)
}

pd.DataFrame(model_scores.items(), columns=["Model", "R2 Score"])
```

## 10. Final Insights

- Living area, location, and construction quality are the strongest drivers of house prices.
- Simple linear models underfit real estate data.
- Polynomial regression captures non-linear pricing behavior.
- Ridge regularization improves generalization by reducing overfitting.
- This approach supports automated valuation models and pricing intelligence systems.
