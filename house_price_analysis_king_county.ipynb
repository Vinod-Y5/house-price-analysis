# House Price Analysis & Prediction – King County, USA

This project analyzes residential house sales data from King County, USA to identify key factors influencing prices and to build regression models for prediction.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression, Ridge
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split

%matplotlib inline
```

## 1. Data Loading & Overview


```python
df = pd.read_csv("housing.csv")
df.head()
```

```python
df.dtypes
```

```python
df.describe()
```

The dataset is loaded and inspected to understand its structure, data types, and basic statistics.


## 2. Data Cleaning & Preprocessing


```python
df.drop(['id', 'Unnamed: 0'], axis=1, inplace=True)
```

```python
df[['bedrooms', 'bathrooms']].isnull().sum()
```

```python
df['bedrooms'] = df['bedrooms'].fillna(df['bedrooms'].mean())
df['bathrooms'] = df['bathrooms'].fillna(df['bathrooms'].mean())
```

```python
df[['bedrooms', 'bathrooms']].isnull().sum()
```

Missing values are handled using mean imputation, and identifier columns are removed.


## 3. Exploratory Data Analysis (EDA)


```python
df['floors'].value_counts()
```

```python
sns.boxplot(x='waterfront', y='price', data=df)
```

```python
sns.regplot(x='sqft_above', y='price', data=df)
plt.ylim(0,)
```

```python
df.select_dtypes(include=np.number).corr()['price'].sort_values(ascending=False)
```

EDA is used to understand price behavior and identify features strongly correlated with house prices.


## 4. Feature Selection & Baseline Regression Models


```python
X = df[['long']]
Y = df['price']

lm = LinearRegression()
lm.fit(X, Y)
lm.score(X, Y)
```

```python
X = df[['sqft_living']]
Y = df['price']

lm.fit(X, Y)
lm.score(X, Y)
```

```python
features = [
    "floors", "waterfront", "lat", "bedrooms", "sqft_basement",
    "view", "bathrooms", "sqft_living15", "sqft_above",
    "grade", "sqft_living"
]

X = df[features]
Y = df['price']

lm.fit(X, Y)
lm.score(X, Y)
```

## 5. Polynomial Regression Using Pipeline


```python
pipe = Pipeline([
    ('scale', StandardScaler()),
    ('polynomial', PolynomialFeatures(include_bias=False)),
    ('model', LinearRegression())
])

pipe.fit(X, Y)
pipe.score(X, Y)
```

## 6. Model Evaluation with Train–Test Split


```python
x_train, x_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.15, random_state=1
)
```

## 7. Ridge Regression & Regularization


```python
ridge = Ridge(alpha=0.1)
ridge.fit(x_train, y_train)
ridge.score(x_test, y_test)
```

```python
pr = PolynomialFeatures(degree=2)

x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.transform(x_test)

ridge.fit(x_train_pr, y_train)
ridge.score(x_test_pr, y_test)
```

## Key Insights

- Living area, location, and quality-related features strongly influence house prices.
- Polynomial and multivariate models outperform simple linear regression.
- Ridge regularization improves generalization on unseen data.
